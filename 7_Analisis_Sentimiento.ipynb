```
import nltk
from google.colab import drive
drive.mount('/content/drive')
```

```
f = open('/content/drive/My Drive/Cursos_Y_Proyectos/Algoritmos_de_clasificacion_de_texto/files/books/negative.review', 'r')
text_negative = f.read()
f.close()
f = open('/content/drive/My Drive/Cursos_Y_Proyectos/Algoritmos_de_clasificacion_de_texto/files/books/positive.review', 'r')
text_positive = f.read()
f.close()
```

```
text_negative.split('\n')[0]
```

```
text_negative.split('#label#:negative\n')[0]
```

```
bag_of_words = []
for line in text_negative.split('#label#:negative\n'):
  for key_value in line.split(' '):
           bag_of_words.append(key_value.split(':')[0]) 
for line in text_positive.split('#label#:positive\n'):
  for key_value in line.split(' '):
           bag_of_words.append(key_value.split(':')[0]) 
fdist = nltk.FreqDist(bag_of_words)
print(f'considerando todo el BOW tendríamos: {len(fdist)} features posibles')
```

```
fdist.plot(50)
```

```
reduced_bow = [w for w in set(bag_of_words) if fdist[w]>50]
```
