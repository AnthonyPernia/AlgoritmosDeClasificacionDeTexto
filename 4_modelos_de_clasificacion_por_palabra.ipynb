```
import nltk, random
nltk.download('names')
from nltk.corpus import names 
import pandas as pd
import string
from sklearn.model_selection import train_test_split
```

```
# definici√≥n de atributos relevantes
def atributos ( palabra):
  return {'ultima_letra': palabra[-1]}
```

```
tagset = [(name,'male' ) for name in names.words('male.txt')] + [(name,'female' ) for name in names.words('female.txt')]
```

```
random.shuffle(tagset)
```

```
tagset[:10]
```

```
fset = [(atributos(name), gender) for (name, gender) in tagset]
```

```
train, test = fset[500:], fset[:500]
```

```
#clasificador naive bayes
classifier = nltk.NaiveBayesClassifier.train(train)
```

```
#le pasamos al clasificador la palabra a testear, recordando que hay que pasarle es los atributos,
#en este caso pasandolo por el metodo atributos() 
classifier.classify(atributos('peter'))
```

```
classifier.classify(atributos('antonio'))
```

```
nltk.classify.accuracy(classifier, test)
```

```
#escogiendo mejores atributos 
abecedario = 'abcdefghijklmnopqrstuvwxyz'
def mas_atributos(nombre):
  atrib = {}
  atrib['primera_letra'] = nombre[0].lower()
  atrib['ultima_letra'] = nombre[-1].lower()
  for letra in abecedario:
    atrib['count({}'.format(letra)] = nombre.lower().count(letra)
    atrib['has({}'.format(letra)] = (letra in nombre.lower())
  return atrib
```

```
mas_atributos('anthony')
```

```
fset = [(mas_atributos(name), gender) for (name, gender) in tagset]
```

```
train, test = fset[500:], fset[:500]
```

```
classifier2 = nltk.NaiveBayesClassifier.train(train)
```

```
nltk.classify.accuracy(classifier2,test)
```

```
def readCSVToArray(path_data, gender):
  data = pd.read_csv(path_data)
  names = [(str(name).lower(), gender) for name in data['name']]
  return names
```

```
male = readCSVToArray('/content/spanish-names-surnames/male_names.csv', 'male')
female = readCSVToArray('/content/spanish-names-surnames/female_names.csv', 'female')
```

```
tagset= male+female
random.shuffle(tagset)
```

```
abecedario = string.ascii_lowercase
```

```
fset = [(mas_atributos(name), gender) for (name, gender) in tagset]
```

```
train, test = train_test_split(fset, test_size=0.2)
```

```
classifier3 = nltk.NaiveBayesClassifier.train(train)
```

```
nltk.classify.accuracy(classifier3,test)
```
